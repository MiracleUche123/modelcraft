```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib
# Clustering and segmentation

# Select the relevant features for clustering and modeling
features = ['Title of course studied', 'What is your highest level of education?', 'Polytechnic/University attended']  # Replace with your chosen features

# Perform K-Means clustering to segment graduates
n_clusters = 3  # Define the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[features])

# Iterate over each cluster and build separate models
for cluster_label in range(n_clusters):
    # Subset the data for the current cluster
    cluster_data = df[df['Cluster'] == cluster_label]

    # Define the target variable for this cluster
    target_column = 'What best describes your current status?'

    # Split the cluster data into training and testing sets
    X = cluster_data[features]
    y = cluster_data[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # Train a Random Forest Classifier for this cluster
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
    rf_classifier.fit(X_train, y_train)

    # Evaluate the model (you can use various metrics)
    accuracy = rf_classifier.score(X_test, y_test)

    # Optionally, you can save the model for each cluster
    model_filename = f'model_cluster_{cluster_label}.pkl'
    joblib.dump(rf_classifier, model_filename)

    # Print cluster-specific results
    print(f'Cluster {cluster_label} - Model Accuracy: {accuracy:.2f}')

# After building models for each cluster, you can use them for prediction or further analysis.
# Accuracy could be improved after efficient data preprocessing
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Check for anomalies

# Select the relevant features for clustering
features = ['Title of course studied', 'What is your highest level of education?', 'Polytechnic/University attended']  # Replace with your chosen features

# Define the number of clusters for K-Means
n_clusters = 3

# Perform K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[features])

# Calculate the distance of each data point to its cluster center
cluster_centers = kmeans.cluster_centers_
df['DistanceToCenter'] = np.linalg.norm(df[features] - cluster_centers[df['Cluster']], axis=1)

# Define a threshold for anomaly detection (you can adjust this threshold)
threshold = 2.0

# Identify anomalies based on the distance to cluster center
anomalies = df[df['DistanceToCenter'] > threshold]

# Display the anomalies
print("Anomalies:")
print(anomalies)


```
