```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define a function to generate and display word clouds
def generate_wordcloud(column_name, title):
    # Combine all text from the specified column into a single string
    text = " ".join(df[column_name].dropna())

    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='viridis').generate(text)

    # Display the word cloud
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title)
    plt.show()

# Create word clouds for the specified columns
generate_wordcloud("Which of these skills/knowledge did your higher education prepare you for?", "Skills/Knowledge Prepared by Higher Education")
generate_wordcloud("As far as you are aware, what was most important to your current employer about your qualification?", "Importance to Current Employer")
generate_wordcloud("What was the most significant reason for deciding to take this job?", "Reason for Taking the Job")
generate_wordcloud("Reason why?", "Additional Reasons for Career Choices")
```
