```python
def converter(df, col):
    try:
        # Fill missing values in the column with the mean
        df[col].fillna(df[col].mean(), inplace=True)

        # Convert the column to integer
        df[col] = df[col].astype(int)
    
    except (ValueError, KeyError):
        print("Error occured, check the column name and data types.")
```

```python
converter(df, "If you answered yes/ongoing to the previous question, what year did you (or will you) complete your NYSC?")
```

```python
converter(df, "How many jobs have you had since graduation including your current one?")
```

```python
def remove_part(data, col):
    
    # Removes the non important parts from a sentence e.g yes: I do - removes :I do.
    
    parts = data[col].split(', ')
    for part in parts:
        if 'yes' in part:
            return part.split('yes: ')[-1]
    return data
```
